```python
import os
import json
drift_dict = {}
for file in os.listdir('/Users/teddy/Documents/TU_e/ThirdYear/Quartile_4/BEP/edbn_BEP/results/driftJSON'):
    with open("/Users/teddy/Documents/TU_e/ThirdYear/Quartile_4/BEP/edbn_BEP/results/driftJSON/{}".format(file), 'r') as f:
        temp = json.load(f)
        #print(file)
        temp_lst = []
        for j in temp.keys():
            temp_lst.append(temp[j])
        drift_dict[file] = temp_lst
            
```
